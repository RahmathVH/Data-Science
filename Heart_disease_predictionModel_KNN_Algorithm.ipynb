```python
#Input  values assigned to x
#Output values assigned to y
x=df.iloc[:,:-1].values  #.values used to print as aaray
x   
y=df.iloc[:,-1].values
y
```

```python
#Split dataset into taining  data and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30) #30% data used as testing 70% used as training
x_train
x_test
y_train
y_test
```

```python
#Normalization method:here we used Standard scalar method 
#Equation:z=(x-u)/s
#u:Mean of training sample
#s:Standard deviation of training sample
#x:each values in data
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()                                #object creation
scalar.fit(x_train)                                    #fit the training data/collect data
x_train=scalar.transform(x_train)                      #normalization
x_test=scalar.transform(x_test)                        #normalization
x_test
```

```python
y_test              #o/p alreday in the y_test
```

```python
#Model creation by using training datas
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)                            #fit i/p and o/p
y_pred=knn.predict(x_test)                          #predict the o/p(y_value) by using test data
y_pred
```

```python
#Performance check:
from sklearn.metrics import confusion_matrix,accuracy_score

#Confusion matrix
mat=confusion_matrix(y_pred,y_test)                        
mat
```

```python
#Accuracy score:
score=accuracy_score(y_pred,y_test)
score
```
